```python
## Берем значения с индексами (0, 1, 2), (0, 1, 3), (0, 2, 3), (1, 2, 3)
for index, value in np.ndenumerate(np.arange(24).reshape(2, 3, 4)):
    i, j, z = index # распаковка кортежа с индексами
    if i < j < z:
        print(value * 10)
```
