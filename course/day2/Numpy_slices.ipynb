```python
import numpy as np
```

```python
# Создаем 3d массив
a3d = np.arange(10, 39)
idx19 = np.where(a3d == 19)
idx29 = np.where(a3d == 29)
a3d = np.delete(a3d, (idx19,idx29))
a3d = a3d.reshape(3, 3, 3)
```

```python
a3d.nbytes, a3d.size, a3d.dtype
```

```python
a3d[1, 2] # a3d[1, 2, :] 
```

```python
a3d[1, 2] # a3d[1, 2, :]
```

```python
a3d[0, :, 1] # 
```

```python
a3d[0, :, 1]
```

```python
a3d[:, 1, 2]  # [15 25 35]
```

```python
a3d[:, 1, 2]
```

```python
a3d[2, ...] #  используем Ellipsis(все остальные элементы)
```

```python
a3d[2, ...] #  используем Ellipsis(все остальные элементы)
```
