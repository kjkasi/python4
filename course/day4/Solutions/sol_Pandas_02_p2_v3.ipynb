```python
student = ["Petrov", "Ivanov", "Sidorov", "Kuznecov", "Popov"]
values = np.random.randint(2, 6, (5, 4))
discip = ["matan", "linal", "matstat", "cs"]

df = pd.DataFrame(values, index=student, columns=discip)
display(df)

# Добавьте еще одного студента и данные по его упеваемости по предметам

df.loc["Vasiliev"] = np.random.randint(2, 6, (4,))
display(df)

# Добавьте предмет physics и оценки по нему
df["physics"] = np.random.randint(2, 6, (6,))
display(df)

# Добавьте признак "mean" ("средний балл"), посчитайте в нем средний балл по каждому студенту.
df["mean"] = df.mean(axis=1)
display(df)

# Посчитайте средний балл по всем предметам, кроме "cs" и "mean"
df["mean_without_cs"] = df[["matan", "linal", "matstat", "physics"]].mean(axis=1)
display(df)

# Удалите предмет "cs"
df.drop("cs", axis=1, inplace=True)
display(df)
```
