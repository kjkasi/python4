```python
# Ваш код
# смотри файл sol_Pandas_02_p2.ipynb
from faker import Faker
from random import randint
```

```python
fake = Faker('ru_RU')  
index_name = [fake.name() for _ in range(5)] 
index_name

estimates = { 'matan': [randint(2, 5) for i in range(5)], 'linal':[randint(2, 5) for i in range(5)],
        'matstat' : [randint(2, 5) for i in range(5)], 'cs': [randint(2, 5) for i in range(5)]
       }
stud = pd.DataFrame(estimates, index=index_name)
stud


```

```python
# В pd.concat() мы кладем датафремы в список
stud_new = pd.concat([stud,
                   pd.DataFrame([[randint(2, 5) for i in range(4)]], index=['Иванов Иван Петрович'], columns=stud.columns)])
stud_new
```

```python
stud_new['physics'] = [randint(2, 5) for i in range(6)]
stud_new
```

```python

stud_new['mean'] = (
   stud_new.mean(axis=1)
)
stud_new
```

```python
# Для каждого студента посчитайте средний балл по всем предметам, кроме "cs" и "mean"

stud_new['mean_2'] = (
(stud_new.sum(axis=1) - (stud_new.loc[:, 'cs'] + stud_new.loc[:, 'mean'])) / (stud_new.shape[1] - 2)
)
stud_new
```

```python
stud_new.drop(['cs'], axis=1, inplace=True)
stud_new
```
